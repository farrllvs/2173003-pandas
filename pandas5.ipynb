**#17 Resampling pada Data Deret Waktu (time series data)**

Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

**Persiapan Data Frame**

```python
n_rows = 365 * 24
n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df
```

**Resampling Data dengan Interval Monthly**

```python
df.resample('M')['col1'].sum().to_frame() #mengubah interval menjadi monthly
```

**Resampling Data dengan Interval Daily**



```python
df.resample('D')['col1'].sum().to_frame() #mengubah monthly based menjadi daily based
```

**#19 Formatting tampilan Data Frame**

**Import Modules**

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

**Persiapan Data Frame**

```python
n_rows = 5
n_cols = 2
cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df['omset'] = df['omset'] * 100_000 #mengkonversi menjadi nilai rupiah
df['operasional'] = df['operasional'] * 10_000
df
```

```python
df.index = pd.util.testing.makeDateIndex(n_rows, freq='D') #menambahkan data tanggal
df = df.reset_index()
df = df.rename(columns={'index':'tanggal'})
df
```

**Melakukan formatting tampilan Data Frame**

```python
formatku = {'tanggal':'{:%d/%m/%y}', 
            'operasional':'Rp {:.2f}',
            'omset':'Rp {:.2f}'} #formatting menjadi tipe data string

laporan = df.style.format(formatku)
laporan # laporan bukan objek dataframe
```

```python
type(laporan)
```

```python
laporan.hide_index() #menghilangkan index (0,1,....)
```

```python
laporan.set_caption('Data Omset dan Operasional') #memberi keterangan pada laporan
```

```python
laporan.highlight_min('omset', color='pink')
laporan.highlight_max('omset', color='lightgreen')

laporan.highlight_min('operasional', color='lightblue')
laporan.highlight_max('operasional', color='grey') #memberi warna pada dataframe
```

**#20 Menggabungkan (merge) Dua Data Frame Secara Berdampingan**

Import Modules

```python
import pandas as pd

print(pd.__version__)
```

```python
d1 = {'col1':[1, 2, 3], 
      'col2':[10, 20, 30]} #membuat 2 dataframe dengan dictionary masing masing
df1 = pd.DataFrame(d1)
df1
```

```python
d2 = {'col3':[4, 5, 6], 
      'col4':[40, 50, 60]}
df2 = pd.DataFrame(d2)
df2
```

```python
df = pd.merge(df1, df2, left_index=True, right_index=True) #melakukan merge antara 2 dataframe
df
```
