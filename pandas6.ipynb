**#21 Agregasi pada Pandas Data Frame dengan agg**

```python
import pandas as pd

print(pd.__version__)
```

**Persiapan Data Frame**

```python
df = pd.read_csv('./Iris.csv') #menggunakan iris dataframe
df
```

**Mengenal groupby() dan Fungsi Agregasi**

```python
df.groupby('Species')['PetalLengthCm'].count().to_frame() #mengelompokan data menggunakan groupby dan menggunakan agregasi mean, median dan count
```

```python
df.groupby('Species')['PetalLengthCm'].mean().to_frame()
```

```python
df.groupby('Species')['PetalLengthCm'].median().to_frame()
```

**Agregasi dengan agg()**

```python
df.groupby('Species')['PetalLengthCm'].agg(['count', 'mean', 'median']) #menggunakan 3 fungsi sekaligus
```

**Agregasi dengan describe()**

```python
df.groupby('Species')['PetalLengthCm'].describe() #menggunakan describe() untuk statistik
```

**#23 Seleksi Baris pada Data Frame dengan Query**

**Import Modules**

```python
import pandas as pd

print(pd.__version__)
```

**Persiapan Data Frame**

```python
d = {'kolom_satu':[1, 2, 3, 4, 5], #pada label tidak menggunakan spasi 
     'kolom dua':[10, 20, 30, 40, 50]} #pada label menggunakan spasi
df = pd.DataFrame(d)
df  
```

**Seleksi baris dengan query()**

```python
df.query('kolom_satu > 2') #query harus berbentuk string dan menggunakan backtick
```

```python
df.query('`kolom dua` > 30') #query harus berbentuk string dan menggunakan backtick
```

**#24 UTC dan konversi zona waktu pada Python Pandas**

**Import Modules**

```python
import pandas as pd

print(pd.__version__)
```

**Persiapan Series**

```python
s = pd.Series(range(1591683521, 1592201921, 3600)) #menggunakan range 1 jam = 3600 detik
s = pd.to_datetime(s, unit='s')
s.head()
```

**Pengaturan zona waktu (time zone)**

```python
s = s.dt.tz_localize('UTC') #localize digunakan utk standar waktu
s.head()
```

```python
s = s.dt.tz_convert('Asia/Jakarta') #convert ke waktu yang diinginkan
s.head()
```

```python
s = s.dt.tz_convert('Australia/Hobart')
s.head()
```
