**Menyertakan Prefix dan Suffix pada seluruh Kolom Data Frame**

Import Modules

**Pemilihan baris atau row selection pada data frame**

**Persiapan Data Frame**

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), 
                  columns = cols) #Jangkauan angka 1-10 dengan dimensi yang ditentukan oleh n_rows dan n_cols
df
```

**Selection dengan operator logika |**

```python
df[(df['A']==1) | (df['A']==3)]
```

**Selection dengan fungsi** isin()

```python
df[df['A'].isin([1, 3])]
```

**Mengenal operator negasi ~**

```python
df[-df['A'].isin([1, 3])]
```

**Konversi tipe data String ke integer pada kolom data frame**

**persiapan data frame**

```python
data = {'col1':['1', '2', '3', 'teks'],
        'col2':['1', '2', '3', '4']} 
        
        
df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

**Konversi menggunakan** astype()

```python
df_x = df.astype({'col2':'int'})
df_x
```

```python
df_x.dtypes
```

**konversi tipe data numerik menggunakan fungsi** to_numeric()

```python
df.apply(pd.to_numeric, errors='coerce')
```

**Pemilihan kolom / colums selection pada data frame berdasarkan tipe datanya**

**persiapan data frame**

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df['bil_pecahan'] = df['bil_pecahan'].astype('float') #mengubah pecahan menjadi float menggunakan function astype()

df.index = pd.util.testing.makeDateIndex(n_rows, freq= 'H') #melakukan interval pada data selama 1 jam sekali
df = df.reset_index()

df['teks'] = list('ABCDE')

df
```

```python
df.dtypes
```

**memilih kolom bertipe data numerik**

```python
df.select_dtypes(include='number')
```

```python
df.select_dtypes(include='float') #memanggil data dengan tipe data float 
```

```python
df.select_dtypes(include='int') #memanggil data dengan tipe data int
```

**Memilih kolom bertipe data string atau object**

```python
df.select_dtypes(include='object') #memanggil data dengan tipe object
```

**memilih kolom bertipe data datetime**

```python
df.select_dtypes(include='datetime') #memanggil data dengan tipe data date
```

**memilih kolom dengan kombinasi tipe data**

```python
df.select_dtypes(include=['number', 'object']) #memanggil data dengan tipe data gabungan antara numerik dan text/string
```
