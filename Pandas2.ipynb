**#05 Membalik Urutan Baris dan Kolom pada Data Frame**

Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan Dataframe

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')
#mengacak nilai random bertipe int

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

Membalik Urutan Kolom

```python
df.loc[:, ::-1] #[x,y], Pada X digunakan untuk mengatur baris dan pada Y digunakan untuk mengatur kolom
```

Membalik Urutan Baris

```python
df.loc[::-1] #[x,y], Pada X digunakan untuk mengatur baris dan pada Y digunakan untuk mengatur kolom
```

Membalik Urutan Baris dan Melakukan Penyesuaian Ulang Index

```python
df.loc[::-1].reset_index(drop=True) #dengan reset_index() kita dapat membalik urutan index dengan angka 0 tetap diatas
```

**#07 Menghapus (drop) Missing Values (NaN) pada Data Frame**

Import Modules

```python
import pandas as pd

print(pd.__version__)
```

Persiapan Data Frame

```python
df = pd.util.testing.makeMissingDataframe().reset_index() #Dataframe ini mengandung missing value
df.head()
```

```python
df = df.rename(columns={'index':'Z'}) #Melakukan rename index menjadi Z
df.head()
```

```python
df_backup = df.copy(deep=True) #melakukan copy/backup menggunakan copy()
```

Menghapus (drop) Setiap Kolom yang Mengandung Missing Values

```python
df = df.dropna(axis='columns') 
df.head() 
```

Menghapus (drop) Setiap Baris yang Mengandung Missing Values

```python
df = df_backup.copy(deep=True)
df = df.dropna(axis='rows') 
df.head()
```

Persentase missing values untuk tiap kolom

```python
df = df_backup.copy(deep=True)

df.isna().mean()*100
#isna() berguna untuk mengumpulkan missing value sedangkan mean() akan menghitung rata2 dari missing value
```

Menghapus (drop) Setiap Kolom yang Mengandung Missing Values Berdasarkan Threshold (ambang batas)

```python
treshold = len(df) * 0.9 #limitasi threshold dapat dilakukan dengan function len()

df = df.dropna(thresh=treshold, axis='columns')
df.head()
```

**#08 Memeriksa Kesamaan Antar Dua Buah Kolom (Series) pada Data Frame**

Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan Data Frame

```python
data = {'A':[15, 15, 18, np.nan, 12], 
        'B':[15, 15, 18, np.nan, 12]}#kolom A dan B memiliki nilai yang sama

df = pd.DataFrame(data)
df
```

Mengenal Pandas Series



*   Series adalah struktur data lain yang dimiliki pandas dengan objek 1-dimensi yang berisi sequence nilai dan berasosiasi dengan label data, yang disebut indeks.
*   Untuk membuat sebuah Series atau kolom, kita bisa membuatnya dari sebuah array dengan memanggil fungsi Series pada pandas.



```python
df['A']
```

```python
type(df['A'])
```

```python
type(df)
```

Memeriksa Kesamaan dengan Operator ==

```python
df['A'] == df['B'] #Kelemahan operator == adalah pengecekan pada NotANumber/NaN kurang baik
```

Memeriksa kesamaan dengan method equals()

```python
df['A'].equals(df['B'])
```

Memeriksa Kesamaan Antar Dua Data Frame

```python
df1 = df.copy(deep=True)

df.equals(df1)
```
