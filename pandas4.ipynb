**#13 Konversi nilai numerik ke dalam kategori pada Data Frame**

**Import Modules**

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

**Persiapan Data Frame**

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

**Pengelompokkan nilai numerik ke dalam beberapa kategori menggunakan cut()**

```python
df['kelompok_usia'] = pd.cut(df['usia'], #kolom yang ingin di cut
                             bins=[0, 18, 65, 99], # batas cut menggunakan list
                             labels=['anak', 'dewasa', 'manula']) #0-18 (anak),18-65(dewasa),65-99(manula)
df
```

**#14 Menggabungkan dua Pandas Data Frame**

**Import Modules**

```python
import pandas as pd
import numpy as np


print(pd.__version__)
print(np.__version__)
```

**Persiapan Data Frame**

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4]) # menghilangkan baris nomor 1 dan 4 dari dataframe
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3]) # menghilangkan baris nomor 0 dan 3 dari dataframe
df2
```

**Menggabungkan Dua Data Frame**

```python
df_inner = pd.merge(df1, df2, how='inner')#menggunakan fungsi merge() secara inner maka akan menghasilkan irisan antara 2 dataframe
df_inner
```

```python
df_outer = pd.merge(df1, df2, how='outer')#menggunakan fungsi merge() secara outer maka akan menghasilkan gabungan antara 2 dataframe
df_outer
```

**#15 Memecah Nilai String dari Suatu Kolom ke Dalam Beberapa Kolom Baru**

**Import Modules**

```python
import pandas as pd

print(pd.__version__)
```

**Persiapan Data Frame**

```python
data = {'nama':['Didi Kempot', 'Glenn Fredly', 'Mbah Surip'], 
        'tempat_kelahiran':['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojokerto, Jawa Timur']}
df = pd.DataFrame(data)
df
```

**Memecah Nama Depan dan Nama Belakang**

```python
df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)#memecah nama depan dan belakang menggunakan fungsi split()
df
```

**Memecah Nama Kota dan Propinsi**

```python
df[['kota', 'propinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)#memecah tempat kelahiran menjadi kota dan propinsi yang dipisahkan dengan koma
df
```

**#16 Menata Ulang Data Frame dengan Mutiple Indexes Menggunakan unstack()**

```python
import pandas as pd

print(pd.__version__)
```

**Persiapan Data Frame**

Dataframe kali ini menggunakan titanic data set

```python
df = pd.read_csv('./data/titanicfull.csv')
df.head()
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean().to_frame()#pengelompokan data menjadi sex dan pclass
```

**Menata Ulang Data Frame dengan Mutiple Indexes**

```python
df.groupby(['sex', 'pclass'])['survived'].mean().unstack()#reshapping data yang memiliki multiple index dengan function unstack()
```
